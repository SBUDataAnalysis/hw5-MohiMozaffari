The multivariate normal, multinormal or Gaussian distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and
covariance matrix. These parameters are analogous to the mean (average or "center") and variance (standard deviation, or "width," squared) of the one-dimensional normal distribution.


Determine whether each sample is correlated, anticorrelated or uncorrelated.

```python
#calculating the corrolation of the samples
print('The correlation of x & y in sample 1 is: ',sample1.corr()['x']['y'])
print('The correlation of x & y in sample 2 is: ',sample2.corr()['x']['y'])
print('The correlation of x & y in sample 3 is: ',sample3.corr()['x']['y'])
```

We know that if correlation is positive, the samples are correlated. If correlation is negative, the samples are anticorrelated. If correlation is zero, the samples are uncorrelated. So, the sample 1 is correlated, the sample 2 is anticorrelated and, the sample 3 is uncorrelated.

## 3d Plot
Plot the joint probability distribution of each sample in 3D. For this you can use 'plot_surface' found in matplotlib library.

If you want to plot using matplotlib, the codes below will come in handy. First line makes sure that your plots are interactive, second line provides color maps.

At the first step try to get the 2D histogram of your data. (Hint: beware of different sizes of arrays!)

```python
bins = 100
#calculating the histograms for sample1
hist1, xedges1, yedges1 = np.histogram2d(sample1['x'], sample1['y'], bins=bins, normed=True)
```

Now you can plot the 3D histogram:

```python
#plotting the histogram for sample1 3D
import matplotlib
matplotlib.rc_file_defaults()


fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection='3d')

#Make data suitable to 3d plotting
X1, Y1 = np.meshgrid(xedges1[:-1], yedges1[:-1])

#Plot the surface
surf = ax.plot_surface(X1, Y1, hist1.T , cmap='viridis', linewidth=0, antialiased=False)

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5)

#adding a title
ax.set_title('Sample 1', fontsize=20)
plt.tight_layout()

plt.show()
```

```python
#plotting the histogram for sample2
hist2, xedges2, yedges2 = np.histogram2d(sample2['x'], sample2['y'], bins=bins, normed=True)

fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection='3d')

#Make data suitable to 3d plotting
X2, Y2 = np.meshgrid(xedges2[:-1], yedges2[:-1])

#Plot the surface
surf2 = ax.plot_surface(X2, Y2, hist2.T , cmap='PiYG_r', linewidth=0, antialiased=False)

# Add a color bar which maps values to colors
fig.colorbar(surf2, shrink=0.5, aspect=5)

#adding a title
ax.set_title('Sample 2', fontsize=20)
plt.tight_layout()

plt.show()
```

```python
#plotting the histogram for sample3
hist3, xedges3, yedges3 = np.histogram2d(sample3['x'], sample3['y'], bins=bins, normed=True)

fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection='3d')

#Make data suitable to 3d plotting
X3, Y3 = np.meshgrid(xedges3[:-1], yedges3[:-1])

#Plot the surface
surf3 = ax.plot_surface(X3, Y3, hist3.T , cmap='rainbow_r', linewidth=0, antialiased=False)

# Add a color bar which maps values to colors
fig.colorbar(surf3, shrink=0.5, aspect=5)

#adding a title
ax.set_title('Sample 3', fontsize=20)
plt.tight_layout()

plt.show()
```


Using the calculated histograms, now write a code to calculate the marginalized PDFs along both axes and then plot them.

```python
#Calculate marginalized distributions
x_marg1 = np.sum(hist1, axis=0)
x_marg2 = np.sum(hist2, axis=0)
x_marg3 = np.sum(hist3, axis=0)
y_marg1 = np.sum(hist1, axis=1)
y_marg2 = np.sum(hist2, axis=1)
y_marg3 = np.sum(hist3, axis=1)
```

```python
x1 = xedges1[:-1]
x2 = xedges2[:-1]
x3 = xedges3[:-1]
y1 = yedges1[:-1]
y2 = yedges2[:-1]
y3 = yedges3[:-1]
```

```python
%matplotlib inline
sns.set()

plt.figure(figsize=(15, 10))
plt.suptitle('Marginalized PDFs')

#1st marginal
plt.subplot(321)
plt.bar(x= x1, height = x_marg1, width= 0.15 , color='magenta')
plt.ylabel('Marginalized PDF')
plt.title('x from sample 1')

#2nd marginal
plt.subplot(322)
plt.bar(x= y1, height = y_marg1, width= 0.15 , color='red')
plt.title('y from sample 1')


#3rd marginal
plt.subplot(323)
plt.bar(x= x2, height = x_marg2, width= 0.15 , color='green')
plt.ylabel('Marginalized PDF')
plt.title('x from sample 2')

#4th marginal
plt.subplot(324)
plt.bar(x= y2, height = y_marg2, width= 0.1 , color='orange')
plt.title('y from sample 2')


#5th marginal
plt.subplot(325)
plt.bar(x= x3, height = x_marg3, width= 0.1 , color='blue')
plt.ylabel('Marginalized PDF')
plt.title('x from sample 3')

#6th marginal
plt.subplot(326)
plt.bar(x= y3, height = y_marg3, width= 0.1 , color='purple')
plt.title('y from sample 3')

plt.tight_layout()
plt.show()
```

## Part 3

Both covariance and correlation measure the relationship and the dependency between two variables. But, correlation values are standardized whereas covariance values are not.
We can obtain the correlation coefficient of two variables by dividing the covariance of these variables by the product of the standard deviations of the same values.

Now select the columns of the train set with numerical data

```python
#creat a new dataframe with numerical values
df_num = df.select_dtypes(include=['int64', 'float64'])
df_num.head()
```

Plot the distribution of all the numerical data

```python
#plotting the histogram of the dataframe with numerical values
fig = plt.figure(figsize=(12, 15))

for i, column in enumerate(df_num.columns, 1):
    plt.subplot(10,4,i)
    sns.histplot(df_num[column],  kde=True, color='magenta')
plt.tight_layout()
plt.show()
```

plot Heatmap for all the remaining numerical data including the 'SalePrice'

```python
#correlation matrix
correlation = df_num.corr()

#plotting the heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation, linewidths=0.1, square=True, cmap='viridis')
plt.title("Heatmap of numerical data")
plt.show()
```

From the distribution of each numerical variables as well as the heatmap you can notice columns that are important and correlated (correlation higher than absolute 0.3) with our target variable 'SalePrice'. select columns where the correlation with 'SalePrice' is higher than |0.3|


```python
#columns where the correlation with 'SalePrice' is higher than |0.3|
corr = correlation[abs(correlation['SalePrice']) > 0.3].SalePrice[:-1].sort_values(ascending=False)
print(corr)
```

Now choose Features with high correlation (higher than 0.5) and plot the correlation of each feature with SalePrice

```python
#plotting cols with high correlation (higher than 0.5) with SalePrice
fig = plt.figure(figsize=(12, 15))

sns.set()
for i, column in enumerate(corr[corr> 0.5].index, 1):
    plt.subplot(5, 2,i)
    sns.scatterplot(x=df_num[column], y=df_num['SalePrice'] , color='#32a852')

plt.tight_layout()
plt.show()
```

Check the NaN of dataframe set by ploting percent of missing values per column and plot the result

```python
#cols with nan value
col_nan = df_num.columns[df_num.isnull().any()]

#number of nan in each column
number_nan = []
number_col = []
for c in col_nan:
    num = df_num[str(c)].isnull().sum()
    number_nan.append(num)
    number_col.append(df_num[str(c)].size)
```

```python
#percent of missing values per column
per = np.array(number_nan)/ np.array(number_col)*100
sns.barplot(x = col_nan, y = per)
plt.title("percent of missing values per column")
plt.ylabel('percent')
plt.show()
```
